```python
#fairly rudimentary zero-crossing on the smoothed first difference
#used a savitzky_golay smoothing algorithm
def peakFind(y):
    y2=savgol.savitzky_golay(y,9,2,1)
    #identify zero crossings as peaks
    threshold = 0.000000019 
    det = y2[:-1]*y2[1:]
    peaks = np.where(det < -threshold)[0]
    return peaks,y2
```

```python
def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
```

### Following block iterates over valid subjects and trials to compute summary statistics

```python
match_pct = np.zeros([max(valid_subjects)+1,17])
blinks_count = np.zeros([max(valid_subjects)+1,17])
val_hist = []
for subj in valid_subjects:
    for trial in range(1,16):
        ex = df[(df['subj']==subj)&(df['trial']==trial)]
        ex = ex[ex['Elapsed_Time']<320]
        peaks,y2 = peakFind(ex['Eyelid_Opening'].values)
        peak_times = ex['Elapsed_Time'][peaks]
        VEOG_peak_times = df2[(df2['Subject']==subj) & (df2['Trial']==trial)]['BlinkTime'].values
        blinks_count[subj,trial]=len(VEOG_peak_times)
        #calc the shift 
        res=400
        theVals = np.zeros(res)
        window_size = 3.5
        shift = np.linspace(1,2*window_size,res)

        for i,val in enumerate(shift):
            error = 0
            for vtime in VEOG_peak_times:
                for ptime in peak_times:
                    #print (vtime-ptime-val)
                    if np.abs(vtime-ptime-val) < 2*window_size:
                        error += (vtime-ptime-val)**2
            theVals[i]+=error

        val = shift[theVals.argmin()]
        val_hist.append(val)
        #find the nearest (shifted) row for the VEOG times
        VEOG_time_index = []
        for atime in VEOG_peak_times:
            VEOG_time_index.append(np.abs(ex['Elapsed_Time'].values - atime-val).argmin())
        
        #determine what percent SES peaks match up with VEOG peaks 
        window = 45
        blinks_vals = list(set(ex['Blink'].values))
        hit = np.zeros(len(VEOG_time_index))
        for i,pindex in enumerate(VEOG_time_index):
            for ind in blinks_vals[1:]:
                theBlink = np.where(ex['Blink'].values==ind)[0]
                low = theBlink[0]-window
                high = theBlink[-1]+window
                #print low,high
                if pindex in range(low,high):
                    #print pindex,low,high
                    hit[i]=1
        match_pct[subj,trial] = hit.sum()/float(len(VEOG_peak_times))
```

```python
#display the best shift value to align peak times
plt.hist(val_hist,25)
plt.show()
```

```python
fwhm = []
for subj in valid_subjects:
    for trial in range(1,16):
        ex = df[(df['subj']==subj)&(df['trial']==trial)]
        ex = ex[ex['Elapsed_Time']<320]
        baseline = ex['Eyelid_Opening'].mean()

        peaks,y2 = peakFind(ex['Eyelid_Opening'].values)
        peak_times = ex['Elapsed_Time'][peaks]
        peak_vals = ex['Eyelid_Opening'].values[peaks]
        VEOG_peak_times = df2[(df2['Subject']==subj) & (df2['Trial']==trial)]['BlinkTime'].values
        len(peak_times),len(VEOG_peak_times)
        
        blah = 5
        peak_ints=[]
        peak_durations=[]
        qual = []

        for peak in peaks:
            peak_ints.append(ex['Eyelid_Opening'].values[peak-blah:peak+blah])
            peak_durations.append(ex['Elapsed_Time'].values[peak-blah:peak+blah])
            qual.append(ex['Eyelid_Quality'].values[peak-blah:peak+blah])
        for i in range(len(peaks)):
            x = peak_durations[i]
            y = baseline-peak_ints[i]
            n = len(x)                          #the number of data
            mean = baseline                #note this correction
            sigma =sum(y*(x-mean)**2)/n  
            try:
                popt,pcov = curve_fit(gaus,x,y,p0=[1,mean,sigma])
                if np.abs(2.355*popt[2])<1.0:
                    fwhm.append(np.abs(2.355*popt[2]))
            except:
                pass

fwhm = np.array(fwhm)
```
